Task 1: Load the User Data

```python
import json
```

```python
# load the data
def load_data(filename):
    with open(filename , "r") as f:
        data = json.load(f)
    return data
```

```python
# check the function
data = load_data("data.json")
data   
```

```python
# check type of data
type(data)
# must be dict
```

```python
# write a function to display users and their connections
def display_users(data):
    for user in data["users"]:
        print(f"Name: {user['name']}")
        print(f"Friends: {user['friends']}")
        print(f"Liked pages: {user['liked_pages']}")
```

```python
def display_users(data):
    print("Users and Their Connections:\n")
    for user in data["users"]:
        print(f"{user['name']} (ID: {user['id']}) - Friends: {user['friends']} - Liked Pages: {user['liked_pages']}")
    print("\nPages:\n")
    for page in data["pages"]:
        print(f"{page['id']}: {page['name']}")

# Load and display the data
data = load_data("data.json")
display_users(data)
```
